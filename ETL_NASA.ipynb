```python
import numpy
import cdflib
import spacepy
import pandas as pd

cdf_data = cdflib.cdf_to_xarray("sol_20220101_v04.cdf")
```

```python
var_time = cdf_data.Epoch.values
valor_tiempo = []
counter_time = 0
for k in var_time:
    valor_tiempo.append(k)
    counter_time +=1
    if counter_time == 25000:
        break
```

```python
valor_magnetico_1 = []
valor_magnetico_2 = []
valor_magnetico_3 = []

counter = 0

for i in cdf_data.BGSE.values:
    valor_magnetico_1.append(i[0])
    valor_magnetico_2.append(i[1])
    valor_magnetico_3.append(i[2])
    counter +=1
    if counter == 25000:
        break
    
```

```python
#data = pd.DataFrame({'Tiempo':valor_tiempo, 'valor_magnetico_1':valor_magnetico_1, 'valor_magnetico_2':valor_magnetico_2, 'valor_magnetico_3':valor_magnetico_3})
```

```python
from math import sqrt
modulo = []
modulo_counter = 0
for i in cdf_data.BGSE.values:
    valor_modulo = sqrt(i[0]**2 + i[1]**2 + i[2]**2)
    modulo.append(valor_modulo)
    modulo_counter += 1
    if modulo_counter == 25000:
        break
modulo
```

```python
#dt = pd.DataFrame({'Tiempo':valor_tiempo, 'modulo':modulo})
```

```python
data_25k = pd.DataFrame({'Tiempo':valor_tiempo, 'valor_magnetico_1':valor_magnetico_1, 'valor_magnetico_2':valor_magnetico_2, 'valor_magnetico_3':valor_magnetico_3, 'modulo':modulo})
```
